# Global Navigation 

For the global navigation part <code>global_utils.py</code>, we implemented the A* algorithm on top of the visibility graph to find the optimal path from the start position to the goal position. After finding the optimal path, we discretize it into a set of rotations and linear movements. These movements are then transformed into a set of motor commands which enables Thymio to reach the goal while avoiding the global (static) obstacles (<strong>Motion</strong> part (<code>motion.py</code>). 

In the following sections, we introduce the procedure and functions we used to do global navigation.

## Modules

Only two main modules used for global navigation, including <strong>numpy</strong> and <strong>shapely</strong>. The <strong>matplotlib</strong> module is also used for plotting the map and the optimal path in the end.

The <strong>numpy</strong> module is famous python library mostly written in C which adds precompiled mathematical and numerical functions and functionalities with great execution speed to python.

<strong>Shapely</strong> is a Python package for manipulation and analysis of planar geometric objects, including points, lines, and polygons. We use <strong>shapely</strong> in the construction of visiblity graph and we will address it in details later.

## Visibility Graph

The aim of having a visibility graph is to construct a set of paths that enables Thymio to move from the start postion to the goal while minimizing the total number of paths. One advantage of the visibility graph is that it is a complete graph. In other words, it means that one can reach any nodes of the graph (particulary the goal) from other nodes. It is also noteworthy that constructing the visiblity graph requies extending the obstacles (in <strong>Vision</strong> part) to prevent Thymio from colliding with them.  

### Nodes
The nodes of the visibility graph are consitsts of the start, goal, and the vertices of all obstacles. The first part of the <strong>vis_graph(obs_coords, start, goal)</strong> function in <code>global_utils.py</code> is dedicated to generate nodes of the graph. We store each node in the following format:<br>
<ul>
<li><em>[node's label (1 for start, total number of nodes for goal), obstacle's number (0 for start, len(obs_coords)+1 for goal), numpy 1d array of x and y coordinates, the number of vertices of the obstacle to which the node belong (0 for start and goal)]</em></li>
</ul>
Some examples are as follows:
<ul>
<li><em>[1, 0, array([350.75,  69.25]), 0]</em></li>
<li><em>[2, 1, array([551, 239]), 4]</em></li>
<li><em>[20, 5, array([637.25, 941.25]), 0]</em></li>
</ul>
The 1<sup>st</sup> element of a node (its label) is useful for plotting and debugging. The 3<sup>rd</sup> element is the x and y coordinates of the node as mentioned previously. The 2<sup>nd</sup> and last elements of the node are useful in defining the edges of the visiblity graph as it is discussed later.

### Edges
For finding the edges of the graph, we loop through all pairs of distinct nodes of the graph and use the procedure described below to check whether an edge is valid or not.
<ol>
<li><strong>Checking whether the chosen nodes is reachable or not:</strong></li>
<par>If some obstacles' vertices are very close to the margins of our environment, we chose to exclude the edges assicated with these nodes in order to avoid having an optimal path which requires Thymio to go very close to our evnironments' boundaries and thus exiting it afterwards. The <strong>is_node_reachable(node, map_margin=20, x_size=720, y_size=1280)</strong> function in <code>global_utils.py</code> in charge of doing this task.</par>
<li><strong>Skipping the obstacles' diagonals:</strong></li>
<par>Since we cannot move through obstacles, we have to exclude the edges corresponding to their diagonals. A diagonal is defined in between two non-consective nodes which belong to the same obstacle (the use of 2<sup>nd</sup> element of a node).</par> 
<li><strong>Adding the obstacles' boundaries to the set of edges</strong></li>
<par>The obstacles' boundaries are among the edges through which Thymio is allowed to move. Obstacles' boundaries are defined between two consecutive nodes which are belong to the same obstacle (the use of 2<sup>nd</sup> element of a node). Regardless, adding the last boundary edge is somewhat tricky. To do that, first we have to find the last vertice of the obstacle corresponding to two consecutive nodes which are belong to two consecutive obstacles (start and goal excluded). Then, by knowing the number of nodes associted with each obstacle (the used of the 3<sup>rd</sup> element of a node), we easily add the last boundary edge of obstacles.</par>
<li><strong>Skipping the collinear edges:</strong></li>
<par>It is possible that sometimes some edges are collinear with each other (example 2 in the end). This happens when an edge contains two vertices of any boundary edges of obstacles. To handle this situation we use <strong>shapely</strong> module. More specifically, first we define a <em>LineString</em> <strong>shapely</strong> object using the two nodes we are looping through to define the edge. Then we call the <strong>create_obs_shape(obs_coords)</strong> function in <code>global_utils.py</code> to transform obstacles' coordinates to a set of <em>Polygon</em> <strong>shapely</strong> object. Afterwards, by using the <strong>.contains</strong> relationship and looping through the polygons, we check whether the edge (<em>LineString</em>) contains any two vertices of the polygons' boundary edges.</par>
<li><strong>Adding the edges which are not crossing any obstacles:</strong></li>
<par>Finally, we should also include the edges which do not cross obstacles. To this, we use the <strong>.crosses</strong> relationship in <strong>shapely</strong> and loop through the polygonns to check whether the edge (<em>LineString</em>) crosses any polygons.</par>
<li><strong>Sorting the edges:</strong></li>
<par>To have a more readble structure, we sort the edges according to the labels of the 1st nodes and then according to the length of the edge</par>
</ol>

The whole point of using the <strong>shapely</strong> module instead of our writing functions to loop through obstacles and check for collinearity and passing through them is that <strong>shapely</strong> uses the GEOS library for all operations which is written in C++ and is highly optimized.

We store each edge in the following format:<br>
<ul>
<li><em>[1st node's label, 1st node's x and y coordinates, 2nd node's label, 2nd node's x and y coordinates, length of the edge]</em></li>
</ul>
Some examples are as follows:
<ul>
<li><em>[1, array([350.75,  69.25]), 2, array([551, 239]), 262.5169042176141]</em></li>
<li><em>[7, array([412, 727]), 8, array([492, 727]), 80.0]</em></li>
<li><em>[18, array([189, 643]), 20, array([637.25, 941.25]), 538.406096733683]</em></li>
</ul>

## A*

As mentioned previously, we use A* to generate the optimal path from start to goal. A* optimizes search by expanding the most promising node chosen to minimize the <strong>f score</strong>.<br>
$f(n) = g(n) + h(n)$
<ul>
<li><strong>G score</strong> or past cost or motion cost is the cost of the cheapest path from start to each node.</li> 
<li><strong>Heuristic</strong> (h) is chosen to be the Euclidean distance of each node to the goal. The <strong>heuristic(nodes)</strong> function in <code>global_utils.py</code> has the role of calculating this heuristic for each node.</li>
</ul>
Our implementation of the A* algorithm, similar to the exercise session 5, is based on the pseudo-code provided in this <a href="https://en.wikipedia.org/wiki/A*_search_algorithm#Pseudocode">link</a>. Sufficient number of comments are written in the <strong>a_star(nodes, edges, h)</strong> function in <code>global_utils.py</code>. It is noteworthy that <strong>a_star</strong> function returns the optimal path from the start node to the goal by calling the <strong>generate_path(path_so_far, current, nodes)</strong> function in <code>global_utils.py</code>.

## Discretizing the Optimal Path

The <strong>discretize_path(start_po, edges, optimal_path)</strong> function in <code>global_utils.py</code> is in charge of dicretizing the optimals path to a set of censecutive rotations and linear movements. We are already calculated the amount of linear movement (the distance in mm) between two nodes (5<sup>th</sup> element of each edge). The <strong>calc_angle(p1, p2)</strong> in <code>global_utils.py</code> calucalates (in radians) the orientation (angle) of a line segement passing through two given nodes.

We store each edge in the following format:<br>
<ul>
<li><em>[1 (rotation) | 2 (linear movement), amount of rotation or liner movement, 1st node's label, 1st node's x and y coordinates, 1st node's orientation, 2nd node's label, 2nd node's x and y coordinates, 2nd node's orientation]</em></li>
</ul>
Some examples are as follows:
<ul>
<li><em>[1, -0.6110922891129029, 1, array([350.75,  69.25]), 1.7350136274493637, 3, array([538, 460]), 1.1239213383364608]</em></li>
<li><em>[2, 86.05230967266364, 3, array([538, 460]), 0.8924256364468225, 10, array([592, 527]), 0.8924256364468225]</em></li>
<li><em>[2, 280.8435952625589, 9, array([619, 661]), 1.505767716516115, 20, array([637.25, 941.25]), 1.505767716516115]</em></li>
</ul>
As it can be seen in the examples above, between each two consecutive nodes of the optimal path, we have a rotation and a linear movement. The rotation is in charge of changing Thymio's previous orientation to the new orientation calculated by <strong>calc_angle</strong> function. It is also noteworthy that we assume that Thymio's orientation in the goal postion is not important; meaning that we want just to reach the goal's x and y coordiants, and the orientation of thymio does not matter in that point.
